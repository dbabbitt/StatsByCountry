```python

max_country_count = pickles_df.country_count.max()
mask_series = (pickles_df.country_count == max_country_count)
master_pickle_name = pickles_df[mask_series].pickle_name.squeeze()
exec_str = f"{master_pickle_name} = s.load_object('{master_pickle_name}')"
exec(exec_str)
mask_series = (pickles_df.pickle_name != master_pickle_name)
for i, r in pickles_df[mask_series].iterrows():
    exec_str = f"{r.pickle_name} = s.load_object('{r.pickle_name}')"
    exec(exec_str)
```

```python

%run ../load_magic/lists.py

mask_series = (pickles_df.pickle_name != master_pickle_name)
for row_index, row_series in pickles_df[mask_series].iterrows():
    left_set = eval(f'set({master_pickle_name}.country_name)')
    right_set = eval(f'set({row_series.pickle_name}.country_name)')
    sd_set = left_set.symmetric_difference(right_set)
    left_list = list(left_set.intersection(sd_set))
    if left_list:
        right_list = list(right_set.intersection(sd_set))
        if right_list:
            typos_df = check_for_typos(left_list, right_list, verbose=False)
            mask_series = (typos_df.max_similarity > 0.0)
            if typos_df[mask_series].shape[0]:
                df = typos_df[mask_series].sort_values(['max_similarity', 'left_item', 'right_item'], ascending=[False, True, True])
                for i, r in df.iterrows():
                    print(f'''
mask_series = ({row_series.pickle_name}.country_name == "{r.right_item}")
{row_series.pickle_name}.loc[mask_series, 'country_name'] = "{r.left_item}"''')
                display(df.rename(columns={'left_item': master_pickle_name, 'right_item': row_series.pickle_name}))
```
