# Create the GDP DataFrames

```python

%run ../load_magic/dataframes.py
import numpy as np

def get_country_gdps_dataframe():
    country_url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
    table_dfs_list = get_wiki_tables(country_url, verbose=True)
    gdp_countries_df = table_dfs_list[0].copy()
    gdp_countries_df.columns = ['country_name', 'country_region', 'imf_estimate', 'imf_year', 'un_estimate', 'un_year', 'wb_estimate', 'wb_year']
    def f(x):
        b_list = str(x).split(']')
        float_str = b_list[-1]
        try:
            float_value = float(float_str)
        except ValueError as e:
            print(f'{type(e)} exception: {str(e).strip()}')
            float_value = np.nan

        return float_value
    for column_name in ['imf_estimate', 'imf_year', 'un_estimate', 'un_year', 'wb_estimate', 'wb_year']:
        gdp_countries_df[column_name] = gdp_countries_df[column_name].map(f)
    s.store_objects(gdp_countries_df=gdp_countries_df)
    
    return gdp_countries_df
```

```python

gdp_countries_df = get_country_gdps_dataframe()
gdp_countries_df.sample(12).T
```

```python

import re

def get_state_gdps_dataframe():
    us_state_url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_GDP'
    table_dfs_list = get_wiki_tables(us_state_url, verbose=True)
    gdp_us_states_df = table_dfs_list[0].copy()
    gdp_us_states_df.columns = ['state_name', 'gdp_millions_usd_2021', 'gdp_millions_usd_2020', 'annual_change_usd',
                                'annual_change_percentage', 'real_gdp_growth_percentage', 'gdp_per_capita_2021', 'gdp_per_capita_2020',
                                'national_precentage_2021', 'national_precentage_2020']
    def f(x):
        float_str = re.sub(r'[^\d\.]+', '', str(x))
        try:
            float_value = float(float_str)
        except ValueError as e:
            # print(f'{type(e)} exception: {str(e).strip()}')
            float_value = np.nan

        return float_value
    for column_name in ['gdp_millions_usd_2021', 'gdp_millions_usd_2020', 'annual_change_usd', 'annual_change_percentage',
                        'real_gdp_growth_percentage', 'gdp_per_capita_2021', 'gdp_per_capita_2020', 'national_precentage_2021',
                        'national_precentage_2020']:
        gdp_us_states_df[column_name] = gdp_us_states_df[column_name].map(f)
    gdp_us_states_df.state_name = gdp_us_states_df.state_name.map(lambda x: str(x).split('*')[0].strip())
    gdp_us_states_df = gdp_us_states_df.drop([0, 52])
    s.store_objects(gdp_us_states_df=gdp_us_states_df)
    
    return gdp_us_states_df
```
