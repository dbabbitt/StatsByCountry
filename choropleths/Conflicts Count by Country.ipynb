```python

page_url = 'https://en.wikipedia.org/wiki/List_of_conflicts_in_the_United_States'
page_soup = nu.get_page_soup(page_url)

# <div class="mw-content-ltr mw-parser-output" lang="en" dir="ltr"><p class="mw-empty-elt">
div_soup = page_soup.find(
    name='div',
    attrs={'class': ['mw-content-ltr', 'mw-parser-output'], 'lang': 'en', 'dir': 'ltr'}
)

# <h2><span class="mw-headline" id="Late_modern_period"><a href="/wiki/Modern_history#Late_modern_period" class="mw-redirect" title="Modern history">
h2s_list = div_soup.find_all(
    name='h2',
)
for h2_soup in h2s_list:
    if not h2_soup.text.startswith('See also'):
        for sibling_soup in h2_soup.next_siblings:
            tag_name = sibling_soup.name
            if tag_name:
                if (tag_name == 'h2'): break
                a_soups_list = sibling_soup.find_all(
                    name='a',
                    attrs={'title': True, 'href': re.compile('^/wiki/')}
                )
                for a_soup in a_soups_list: print(a_soup['title'])
```
